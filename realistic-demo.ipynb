Establish true overdensity field and samples thereof

```python
# frequency = 100.

# deltath = np.sin(zs*frequency)
zs = []
rhos = []
with open('onerun.txt', 'rb') as csvfile:
    filereader = csv.reader(csvfile, delimiter=',')
    for row in filereader:
        zs.append(float(row[0]))
        rhos.append(float(row[1]))
zs = np.array(zs)
rhos = np.array(rhos)
minz = min(zs)
maxz = max(zs)
zrange = maxz-minz

deltaint = np.log(rhos)/np.log(np.mean(rhos))

plt.plot(zs,deltaint,c='k')
plt.xlabel('redshift')
plt.ylabel('overdensity')
plt.savefig('truedensity.png')
```

```python
#scipy.ndimage,filters.convolve(zs,deltath)
b = sp.signal.gaussian(100, 30)
deltath = sp.ndimage.filters.convolve1d(deltaint, b/b.sum())

plt.plot(zs,deltath,c='k')
plt.xlabel('redshift')
plt.ylabel('overdensity')
plt.savefig('smoothdensity.png')
```

```python
deltaobs = []
zobs = []
nsamps = 100
for i in xrange(len(zs)):
    if i%30 == 0:
        deltaobs.append(deltath[i])
        zobs.append(zs[i])
deltaobs = np.array(deltaobs)
zobs = np.array(zobs)

errconst = 0.1*(max(deltaobs)-deltaobs)

deltaerr = np.sqrt(errconst*np.abs(deltaobs))
delta = deltaobs + deltaerr * np.random.randn(len(zobs))
plt.plot(zobs,delta,'bo')
plt.errorbar(zobs, delta, deltaerr, color='b', markersize=10)#, fmt='k.', markersize=10, label=u'Observations',
                 #ecolor='gray')
plt.plot(zs,deltath,c='k')
plt.xlabel('redshift')
plt.ylabel('overdensity')
plt.savefig('al-observeddensity.png')
```

GP fit to true overdensity field

```python
kernel_param = 0.001
kernel = ExpSquaredKernel(kernel_param)#chosen by magic

gp = george.GP(kernel)

gp.compute(zobs, deltaerr)

ztest = np.linspace(min(zobs), max(zobs), 500)

mu, cov = gp.predict(delta, ztest)
std = np.sqrt(np.diag(cov))

nsamps = 10
toplot = np.random.multivariate_normal(mu,cov,nsamps)
plt.plot(zobs,delta,'bo')
plt.errorbar(zobs, delta, deltaerr, color='b')
for s in xrange(nsamps):
    plt.plot(ztest,toplot[s],c='r',alpha=5./nsamps)
plt.plot(zs,deltath,c='k')
plt.xlabel('redshift')
plt.ylabel('overdensity')
plt.savefig('al-gptoobs.png')
```

Generate probabilities of being Lyman alpha for each line

```python
pop_frac = 0.5#fraction of Ly-a lines expected

norm_delta_err = np.array([min((delta[n]+1.)/2.,1.) for n in xrange(len(delta))])
norm_delta_err = np.array([max(norm_delta_err[n]*(1.-pop_frac),0.) for n in xrange(len(delta))])

def make_one_prob(err):
    p = err*np.random.rand(1)
    return p
def make_probs(deltaerr):
    N = len(deltaerr)
    ps = [make_one_prob(deltaerr[n]) for n in xrange(N)]
    return ps

npix = len(zobs)
p = make_probs(norm_delta_err)
```

Make catalogs of survey instantiations

```python
def add_one_to_cat(y,xt,p):
    if np.random.uniform() > p:
        return xt
    return minz+zrange*np.random.rand(1)#or bias this
def make_cat(ys,xts,ps):
    output = []
    for n in xrange(len(ps)):
        output.append(add_one_to_cat(ys[n],xts[n],ps[n]))
    return np.array(output)

ncats = 100
obszs = [make_cat(delta,zobs,p) for n in xrange(ncats)]

print(len(obszs[0]),len(zobs))
for xo in obszs:
    plt.plot(xo,delta,'ro',alpha=1./5.)
    #plt.errorbar(xo, delta, deltaerr, color='r', markersize=10)
plt.plot(zobs,delta,'bo')
plt.errorbar(zobs, delta, deltaerr, color='b')
plt.plot(zs,deltath,c='k')
plt.xlabel('redshift')
plt.ylabel('overdensity')
plt.savefig('alinstantiations.png')
```

Fit GP to instantiations of catalog

```python
gp_obss = [george.GP(kernel) for c in range(ncats)]

mu_obss,cov_obss,std_obss = [],[],[]
for c in range(ncats):
    gp_obss[c].compute(obszs[c], deltaerr)
    mu, cov = gp_obss[c].predict(delta, ztest)
    std = np.sqrt(np.diag(cov))
    mu_obss.append(mu)
    cov_obss.append(cov)
    std_obss.append(std)
mu_obss = np.array(mu_obss)
cov_obss = np.array(cov_obss)
std_obss = np.array(std_obss)
```

Plot the results!

```python
x_pred = ztest
y_pred = np.median(mu_obss,axis=0)#np.sum(mu_obss,axis=0)/ncats#mu
sigma = np.sqrt(np.sum(std_obss**2,axis=0)/ncats)#std
print(np.shape(sigma))
y_preds = mu_obss
sigmas = std_obss

def plot_results_all(X, y, dy, x_pred, y_pred, sigma, y_preds, sigmas):
    fig = plt.figure(figsize=(8, 6))
    plt.plot(zs,deltath,c='k',label='true distribution')
    #plt.plot(zs, delta, 'k:', label=u'$f(x) = \sin(100*x)$')
    plt.errorbar(X.ravel(), y, dy, fmt='k.', markersize=10, label=u'Observations',
                 ecolor='gray')
    plt.plot(x_pred, y_pred, 'b-', label=u'Prediction')
    plt.fill(np.concatenate([x_pred, x_pred[::-1]]),
             np.concatenate([y_pred - sigma,
                            (y_pred + sigma)[::-1]]),
             alpha=.5, fc='b', ec='None', label='1$\sigma$ confidence interval')
    for xo in obszs:
        plt.plot(xo,delta,'ko',alpha=1./5.)
    for c in xrange(ncats):
        plt.plot(x_pred, y_preds[c], 'r-',alpha=10./ncats)
        plt.fill(np.concatenate([x_pred, x_pred[::-1]]),
             np.concatenate([y_preds[c] - sigmas[c],
                            (y_preds[c] + sigmas[c])[::-1]]),
             alpha=2./ncats, fc='r', ec='None')
    plt.xlabel('$z$')
    plt.ylabel('overdensity')
    plt.legend(fontsize='small',loc='lower right');
    plt.savefig('al-final_results.png')
    
plot_results_all(zobs, delta, deltaerr, x_pred, y_pred, sigma, y_preds, sigmas)
```
